```
# Download YouTube video using yt-dlp
import yt_dlp

def download_youtube_video_yt_dlp(video_url, output_path):
    try:
        ydl_opts = {
            'format': 'best',
            'outtmpl': f"{output_path}/%(title)s.%(ext)s"
        }

        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(video_url, download=True)
            print(f"Downloaded: {info['title']} to {output_path}")
            return {"title": info['title'], "views": info['view_count'], "duration": info['duration']}
    except Exception as e:
        print(f"Error: {e}")
        return None
```

```
# Test with yt-dlp
video_url = "https://www.youtube.com/watch?v=syFZfO_wfMQ"  # Replace with your desired URL
metadata_vid = download_youtube_video_yt_dlp(video_url, output_video_path)
print(metadata_vid)
```

```
# Extract frames and audio from video
def video_to_images(video_path, output_folder):
    clip = VideoFileClip(video_path)
    clip.write_images_sequence(
        os.path.join(output_folder, "frame%04d.png"), fps=0.2
    )
```

```
def video_to_audio(video_path, output_audio_path):
    clip = VideoFileClip(video_path)
    audio = clip.audio
    audio.write_audiofile(output_audio_path)
```

```
# Convert audio to text using Hugging Face's Whisper
def audio_to_text(audio_path):
    recognizer = sr.Recognizer()
    audio = sr.AudioFile(audio_path)
    with audio as source:
        audio_data = recognizer.record(source)
        try:
            # Use Hugging Face's Whisper model
            pipe = pipeline("automatic-speech-recognition", model="openai/whisper-small")
            # Add return_timestamps=True for long-form generation
            text = pipe(audio_data.get_wav_data(), return_timestamps=True)
        except sr.UnknownValueError:
            print("Speech Recognition could not understand audio")
    return text["text"] # Still return text["text"] to keep the original output format
```

**Processing the video**

```
# Process video
def process_video(video_url, output_video_path, output_folder, output_audio_path):
    metadata_vid = download_youtube_video_yt_dlp(video_url, output_video_path)
    filepath = os.path.join(output_video_path, f"{metadata_vid['title']}.mp4")

    video_to_images(filepath, output_folder)
    video_to_audio(filepath, output_audio_path)
    text_data = audio_to_text(output_audio_path)

    with open(os.path.join(output_folder, "output_text.txt"), "w") as file:
        file.write(text_data)

    os.remove(output_audio_path)
    print("Audio file removed")

    return text_data
```

```
# Process the video
text_data = process_video(video_url, output_video_path, output_folder, output_audio_path)
```

```
# Load Hugging Face models for text and image embeddings
text_model_name = "bert-base-uncased"
image_model_name = "google/vit-base-patch16-224"

text_tokenizer = AutoTokenizer.from_pretrained(text_model_name)
text_model = AutoModel.from_pretrained(text_model_name)

image_feature_extractor = AutoFeatureExtractor.from_pretrained(image_model_name)
image_model = AutoModel.from_pretrained(image_model_name)
```

```
# Generate embeddings
def get_text_embedding(text):
    inputs = text_tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        outputs = text_model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).numpy()

def get_image_embedding(image):
    inputs = image_feature_extractor(images=image, return_tensors="pt")
    with torch.no_grad():
        outputs = image_model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).numpy()
```

```
# Set up LanceDB for vector storage
text_store = LanceDBVectorStore(uri="lancedb", table_name="text_collection")
image_store = LanceDBVectorStore(uri="lancedb", table_name="image_collection")

storage_context = StorageContext.from_defaults(
    vector_store=text_store,
    image_store=image_store
)
```

```
# Load documents
documents = SimpleDirectoryReader(output_folder).load_data()
```

```
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# Use Hugging Face BERT-based model for text embeddings
embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-MiniLM-L6-v2")

index = MultiModalVectorStoreIndex.from_documents(
    documents, storage_context=storage_context, embed_model=embed_model
)

```

```
# Create retriever engine
retriever_engine = index.as_retriever(similarity_top_k=1, images_similarity_top_k=10)
```

```
import os
import matplotlib.pyplot as plt
from PIL import Image
from llama_index.core.response.notebook_utils import display_source_node
from llama_index.core.schema import ImageNode

# Define retrieval function
def retrieve(retriever_engine, query_str):
    retrieval_results = retriever_engine.retrieve(query_str)

    retrieved_images = []
    retrieved_text = []
    for res_node in retrieval_results:
        if isinstance(res_node.node, ImageNode):
            retrieved_images.append(res_node.node.metadata["file_path"])
        else:
            display_source_node(res_node, source_length=200)
            retrieved_text.append(res_node.text)

    return retrieved_images, retrieved_text

# Define function to plot more images (up to 10)
def plot_images(images_path, max_images=10):
    images_shown = 0
    num_cols = 5  # Display images in rows of 5
    num_rows = (len(images_path) // num_cols) + 1

    plt.figure(figsize=(18, 9))
    for img_path in images_path:
        if os.path.isfile(img_path):
            image = Image.open(img_path)

            plt.subplot(num_rows, num_cols, images_shown + 1)
            plt.imshow(image)
            plt.xticks([])
            plt.yticks([])
            plt.title(f"Image {images_shown + 1}")

            images_shown += 1
            if images_shown >= max_images:
                break

    plt.tight_layout()
    plt.show()  # Display images

# Define query
query_text = "What is the main topic of the video?"

# Retrieve text and images
img, text = retrieve(retriever_engine, query_text)

# Print extracted text
print("Extracted Text:")
for i, t in enumerate(text):
    print(f"Text {i+1}: {t}")

# Plot retrieved images
plot_images(img, max_images=10)
```
